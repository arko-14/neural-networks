```python
from sklearn.model_selection import train_test_split
```

```python
labels = df['label'].values
X = df.drop('label', axis=1).values.astype(np.float32)
```

```python
X /= 255.0

# One-hot encode the labels
def one_hot_encode(labels, num_classes=10):
    m = labels.shape[0]
    one_hot = np.zeros((m, num_classes))
    labels = labels.astype(int)
    one_hot[np.arange(m), labels] = 1
    return one_hot

y_encoded = one_hot_encode(labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)


X_train = X_train.T
X_test = X_test.T
y_train = y_train.T
y_test = y_test.T

print("Training data shape:", X_train.shape)
print("Training labels shape:", y_train.shape)
print("Testing data shape:", X_test.shape)
print("Testing labels shape:", y_test.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(Z):
    """ReLU activation function"""
    return np.maximum(0, Z)

def softmax(Z):
    """Softmax activation function"""
    expZ = np.exp(Z - np.max(Z, axis=0, keepdims=True))
    return expZ / np.sum(expZ, axis=0, keepdims=True)

def initialize_parameters(input_size, hidden_size, output_size):
    """Initialize neural network parameters with He initialization for ReLU"""
    W1 = np.random.randn(hidden_size, input_size) * np.sqrt(2 / input_size)
    b1 = np.zeros((hidden_size, 1))
    W2 = np.random.randn(output_size, hidden_size) * 0.01
    b2 = np.zeros((output_size, 1))
    return {"W1": W1, "b1": b1, "W2": W2, "b2": b2}

def forward_propagation(X, parameters):
    """Forward propagation with ReLU in hidden layer"""
    W1, b1, W2, b2 = parameters["W1"], parameters["b1"], parameters["W2"], parameters["b2"]

    # First layer with ReLU
    Z1 = np.dot(W1, X) + b1
    A1 = relu(Z1)

    # Output layer with softmax
    Z2 = np.dot(W2, A1) + b2
    A2 = softmax(Z2)

    cache = {"Z1": Z1, "A1": A1, "Z2": Z2, "A2": A2}
    return A2, cache

def compute_cost(A2, Y):
    """Compute cross-entropy loss"""
    m = Y.shape[1]
    cost = -np.sum(Y * np.log(A2 + 1e-8)) / m
    return cost

def backward_propagation(X, Y, parameters, cache):
    """Backward propagation with ReLU derivative"""
    m = X.shape[1]
    W1, W2 = parameters["W1"], parameters["W2"]
    A1, A2 = cache["A1"], cache["A2"]

    # Softmax derivative for output layer
    dZ2 = A2 - Y
    dW2 = np.dot(dZ2, A1.T) / m
    db2 = np.sum(dZ2, axis=1, keepdims=True) / m

    # ReLU derivative for hidden layer
    dZ1 = np.dot(W2.T, dZ2) * (A1 > 0)
    dW1 = np.dot(dZ1, X.T) / m
    db1 = np.sum(dZ1, axis=1, keepdims=True) / m

    grads = {"dW1": dW1, "db1": db1, "dW2": dW2, "db2": db2}
    return grads

def update_parameters(parameters, grads, learning_rate):
    """Update network parameters"""
    for key in parameters:
        parameters[key] -= learning_rate * grads["d" + key]
    return parameters

def predict(X, parameters):
    """Make predictions"""
    A2, _ = forward_propagation(X, parameters)
    return np.argmax(A2, axis=0)

def plot_evolution(distribution_history, checkpoints):
    """Plot evolution of class predictions"""
    distribution_history = np.array(distribution_history)
    num_classes = distribution_history.shape[1]

    plt.figure(figsize=(10, 6))
    for i in range(num_classes):
        plt.plot(checkpoints, distribution_history[:, i], marker='o', label=f'Class {i}')
    plt.xlabel('Epoch')
    plt.ylabel('Number of Predictions')
    plt.title('Evolution of Class Prediction Counts Over Training')
    plt.legend()
    plt.grid(True)
    plt.show()

def train_neural_network(X_train, y_train, X_test, y_test, hidden_size=64, learning_rate=0.1, epochs=1000):

    input_size = X_train.shape[0]
    output_size = y_train.shape[0]
    parameters = initialize_parameters(input_size, hidden_size, output_size)

    distribution_history = []
    checkpoints = []

    for epoch in range(epochs):
        # Forward propagation
        A2, cache = forward_propagation(X_train, parameters)

        # Compute cost
        cost = compute_cost(A2, y_train)

        # Backward propagation
        grads = backward_propagation(X_train, y_train, parameters, cache)

        # Update parameters
        parameters = update_parameters(parameters, grads, learning_rate)

        # Periodic reporting
        if epoch % 100 == 0:
            predictions = predict(X_train, parameters)
            accuracy = np.mean(predictions == np.argmax(y_train, axis=0)) * 100
            print(f"Epoch {epoch}: Cost {cost:.4f}, Training Accuracy: {accuracy:.2f}%")

            # Record the class counts for this checkpoint
            counts = np.bincount(predictions, minlength=10)
            distribution_history.append(counts)
            checkpoints.append(epoch)

    # Final evaluation
    test_predictions = predict(X_test, parameters)
    test_accuracy = np.mean(test_predictions == np.argmax(y_test, axis=0)) * 100
    print(f"Final Test Accuracy: {test_accuracy:.2f}%")

    # Plot evolution of prediction counts over training
    plot_evolution(distribution_history, checkpoints)

    return parameters

trained_parameters = train_neural_network(X_train, y_train, X_test, y_test)
```

```python
prediction = predict(X_test[:, 5].reshape(-1, 1), trained_parameters)
print(prediction) # Prints the prediction
```

```python
test_predictions = predict(X_test, trained_parameters)

test_accuracy = np.mean(test_predictions == np.argmax(y_test, axis=0)) * 100
print(f"Test Accuracy: {test_accuracy:.2f}%")

print("Predictions for first 10 test samples:", test_predictions[:10])
```

```python


import matplotlib.pyplot as plt
import numpy as np



true_labels = np.argmax(y_test, axis=0)

# Create a confusion matrix
confusion_matrix = np.zeros((10, 10), dtype=int)
for i in range(len(test_predictions)):
    confusion_matrix[true_labels[i], test_predictions[i]] += 1

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
plt.imshow(confusion_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(10)
plt.xticks(tick_marks, tick_marks)
plt.yticks(tick_marks, tick_marks)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

# Add labels to the plot
for i in range(10):
    for j in range(10):
        plt.text(j, i, str(confusion_matrix[i, j]), ha='center', va='center', color='white' if confusion_matrix[i, j] > confusion_matrix.max() / 2 else 'black')

plt.show()
```
